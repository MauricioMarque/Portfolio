## Use Prophet Model for Easy Python Forecasting

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
df=pd.read_csv('forecastdata.csv',delimiter=';')
```

```python
df.head()
```

```python
df.info() 
```

```python
#  let's change the Date Type from object to date
df['Date']=pd.to_datetime(df['Date'])
```

```python
ts=df
ts.columns=['ds','y']
ts.head()
```

```python
ts.set_index('ds')
```

```python
df.set_index('ds').plot()
```

```python
model=Prophet().fit(ts)
future=model.make_future_dataframe(60)
forecast=model.predict(future)
forecast.head()
```

```python
model.plot(forecast);
```

```python
model.plot_components(forecast);
```

```python
from prophet.plot import add_changepoints_to_plot
```

```python
fig=model.plot(forecast)
add_changepoints_to_plot(fig.gca(),model,forecast)
```
